#### Merging the two GeoJSONs

```python
countries_json = open("../Resources/countries.geo.json").read()
us_json = open("../Resources/us-states.json").read()

countries = json.loads(countries_json)
us = json.loads(us_json)
```

```python
# pprint(countries)
```

```python
inner_json = us["features"]
# inner_json
```

```python
for item in inner_json:
    countries["features"].append(item)

# countries["features"]
```
